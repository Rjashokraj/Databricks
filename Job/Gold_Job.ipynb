# creating Gold tables
# Gold layer will contain business insight tables

```python
df = spark.table("db_test.silver.movies")
display(df)
```

# Revenue by Industry

```python
from pyspark.sql.functions import sum

Revenue = df.groupBy("industry") \
    .agg(sum("revenue").alias("total_revenue"))

Revenue.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("db_test.gold.revenue_by_industry")

display(Revenue)
```

# Revenue by Year

```python
df_year = df.groupBy("release_year") \
    .agg(sum("revenue").alias("total_revenue"))

df_year.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("db_test.gold.revenue_by_year")
```

# Top Rated Movies

```python
from pyspark.sql.functions import desc

df_top = df.orderBy(desc("imdb_rating")) \
    .limit(10)

df_top.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("db_test.gold.top_movies")
```
